```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#cargas datos en dataframe
datos = pd.read_csv("https://raw.githubusercontent.com/CodigoMaquina/code/main/datos/ingreso.csv")

datos.head()
```

```python
import matplotlib.pyplot as plt

plt.ylabel("Ingreso($)")
plt.xlabel("Promedio de horas semanales trabajadas")
plt.scatter(datos["horas"], datos["ingreso"],color="pink")
plt.show()

```

```python
import matplotlib.pyplot as plt

plt.ylabel("Ingreso($)")
plt.xlabel("Promedio de horas semanales trabajadas")
plt.hexbin(datos["horas"], datos["ingreso"], gridsize=50, cmap='pink')
plt.show()

```

```python
#crear modelo de regresion lineal simple, para obtener el ingreso en base a horas trabajadas
```

```python
from sklearn import linear_model

regresion= linear_model.LinearRegression()  #Solicitud de modelo de regresion lineal

#los datos de entrada deben de tener un formato especifico
datos["horas"]
```

```python
#poner datos en vector, necsitamos un vector que tenga multiples vectores adentro

datos["horas"].values.reshape((-1, 1)) #solicitud de vectores de un solo 1 elemento
```

```python
#Tenemos 1 vector y dentro del vector los elementos estn dentro de 1 vector
```

```python
horas = datos["horas"].values.reshape((-1,1))

modelo=regresion.fit(horas, datos["ingreso"])

#Modelo permite predecir el ingreso de salario por horas trabajadas
```

```python
print("Interseccion(b)", modelo.intercept_)
print("Pendiente(m)", modelo.coef_)

```

```python
#resultado: para cada hora trabajada por los empleados , aumenta el salario caso 3000 pesos.

# ahora tenemos una ecuacion de la linea de recta y manualmente podriamos obtener cual es el potencial salario
# sin embargo el modelo tiene un metodo que permite realizar predicciones.
```

```python
entrada = [[39],[40],[43],[44]]
modelo.predict(entrada)
```
