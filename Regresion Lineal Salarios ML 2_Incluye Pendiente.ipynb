```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset=pd.read_csv('https://raw.githubusercontent.com/CodigoMaquina/code/main/datos/ingreso.csv')
```

```python
dataset
```

```python
dataset.shape
```

```python
#se tienen 100 observaciones
```

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,1].values
```

```python
# En el caso de X, el código recorre todas las filas y descarta la última columna.
# En el caso de Y, considera todas las filas y solo considera la última columna.
```

```python
# se crearán 4 variables:

#     X_train: entrenamiento para variable x.
#     X_test: test para variable x.
#     Y_train: entrenamiento para variable y.
#     Y_test: test para variable y.


# A cada uno de estos datos los vamos a dividir en entrenamiento y 
# validación con el método train_test_split (como ya se ha importado en el paso 1)
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
X_train
```

```python
regressor= LinearRegression()
regressor.fit(X_train,Y_train)

#Modelo listo
```

```python

#Visualizacion,revision de entrenamiento
viz_train = plt
viz_train.scatter(X_train, Y_train,color='pink')
#viz_train.hexbin(X_train, Y_train, gridsize=25, cmap='pink')
viz_train.plot(X_train,regressor.predict(X_train),color='blue')
viz_train.title('Ingreso vs Horas trabajadas')
viz_train.xlabel('Promedio de horas trabajadas')
viz_train.ylabel('Ingreso Aproximado')
viz_train.colorbar(label='counts')
viz_train.show()

```

```python
regressor= LinearRegression()
regressor.fit(X_train,Y_train)

#Visualizacion de pruebas
viz_test = plt
viz_test.scatter(X_test, Y_test,color='pink')
#viz_train.hexbin(X_train, Y_train, gridsize=25, cmap='pink')
viz_test.plot(X_test,regressor.predict(X_test), color='blue')
viz_train.title('Salario vs Horas trabajadas')
viz_train.xlabel('Promedio de horas trabajadas')
viz_train.ylabel('ingreso')
viz_test.colorbar(label='counts')
viz_test.show()
```

```python
#score
regressor.score(X_test,Y_test)
```
